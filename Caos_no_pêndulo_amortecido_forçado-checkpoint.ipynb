O objetivo deste projeto é simular um pêndulo físico (amortecido e foçado), observar seu comportamento em regimes caóticos e não-caóticos, e por fim analisar como se dá a transição de um regime para o outro a partir de um diagrama de bifurcação. **Importando as bibliotecas:**

```python
from numpy import sin, array, linspace, pi
from pylab import plot, show, figure,xlabel, ylabel, xlim, ylim, title, grid
```

**Equação diferencial para o pêndulo físico:**

Considere um pendulo formado por uma barra de massa despresível e comprimento inextensível $l$, com um dos seus extremos preso em um ponto fixo e o outro preso a uma massa $m$ pontual. O angulo formado entre a barra e o eixo vertical é $\theta$, e a aceleração da gravidade $g$. Para um pêndulo simples, sem atrito e força externa, a força resultante sobre a massa na coordenada $\theta$ é:

$$F_{\theta} = -mg\ sin(\theta)$$

Se $F_\theta = m\dfrac{{d^2}s}{dt^2}$, onde $s$ é o arco da circunferencia definido por $s = \theta l$, então:
$$\dfrac{{d^2}\theta}{dt^2} = -\dfrac{g}{l}sin(\theta)$$

Para adicionar atrito a equação vamos considerar que, independente da fonte do atrito, ele é diretamente proporcional a velocidade angular do pêndulo $w = \dfrac{d\theta}{dt}$, assumindo assim a forma $-q\dfrac{d\theta}{dt}$, onde $q$ é um fator relacionado a intensidade do atrito. Assim, a equação para o pêndulo simples com atrito fica:
$$\dfrac{{d^2}\theta}{dt^2} = -\dfrac{g}{l}sin(\theta) -q\dfrac{d\theta}{dt}$$

Por fim, adicionamos uma força externa senoidal da forma $F_{D}sin(Dt)$, onde $F_D$ é um fator que indica a intensidade da força, e $D$ é sua frequência angular. Logo, a equação diferencial que descreve o movimento de um pendulo físico assume a forma:

$$\dfrac{{d^2}\theta}{dt^2} = -\dfrac{g}{l}sin(\theta) -q\dfrac{d\theta}{dt} + F_{D}sin(Dt)$$

O método computacional usado para resolver esta equação diferencial do segundo gral, é transforma-la em duas equações do primeiro grau e resolve-las. Desta forma, a solução para o pêndulo físico pode ser encontrada resolvendo:

$$\dfrac{d\theta}{dt} = w$$

$$\dfrac{{d}w}{dt} = -\dfrac{g}{l}sin(\theta) -q\dfrac{d\theta}{dt} + F_{D}sin(Dt)$$

**Criando uma função com as EDOs:**

Esta função recebe o valor atual da velocidade angular $w$ e do angulo $\theta$, por meio da lista $r$, além do tempo $t$, e os usa para calcular a aceleração angular atual do pêndulo. 

```python
def f(r,t):
    w,the = r[0],r[1]
    dthe = w
    dw = -(g/l)*sin(the)-q*dthe+Fd*sin(D*t)
    
    return array([dw,dthe])
```

**Criando uma função para o método Runge Kutta:**

O método Runge Kutta de quarta ordem é um método computacional bastante preciso, usado para resolver equações diferenciais. Na verdade, ele é um conjunto de métodos para resolver EDOs, que incluem por exemplo o método de Euler. Seu funcionamento basicamente consiste em receber dados iniciais de uma função e dizer onde ela estará no passo seguinte.

```python
def RK4(r,t):
    k1 = h*f(r,t)
    k2 = h*f(r+k1/2,t+h/2)
    k3 = h*f(r+k2/2,t+h/2)
    k4 = h*f(r+k3,t+h)    
    return r + (1/6)*(k1+2*k2+2*k3+k4)
```

**Resolvendo o pêndulo físico:**

Primeiro vamos testar a equação para o pendulo físico, plotando o valor de $\theta$ em função do tempo ($\theta-t$) e o espaço de fase do pêndulo ($w-\theta$). Para facilitar a visualização dos resultados vamos conter a variável $\theta$ no intervalo $]-\pi,\pi]$, adicionando ou subtraindo $2\pi$ caso ele venha a sair do intervalo. O pêndulo vai ser simulado entre os tempos $a$ e $b$, a simulação vai ser dividida em $N$ partes e o passo de tempo vai ser $h$. Além disso os valores atuais de $\theta$ e $w$ vão estar na lista $r$, sendo estes e os tempos armazenados nas listas $X$, $Y$ e $T$, para serem plotados posteriormente. Neste caso, as constantes do sistema vão assumir os valores $g = 9.8$, $l = 1.0$, $F_D = 0.2$, $D = 2.0$ e $q = 1.0$, já as condições iniciais do pêndulo vão ser: $\theta(0) = 0.2$ e $w(0) = 0$.

```python
g = 9.8
l = 1.0
D = 2.0
q = 1.0
Fd = 0.2
a = 0.0
b = 20.0
N = 5*10**3
h = (b-a)/N

w1,the1 = 0,0.2
r1 = array([w1,the1])
X1 = []
Y1 = []
T = []

for t in linspace(a,b,N):
    if r1[1]>pi:
        r1[1]-=2*pi
    if r1[1]<=-pi:
        r1[1]+=2*pi
    X1.append(r1[0])
    Y1.append(r1[1])
    T.append(t)
    r1 = RK4(r1,t) 
    
title("Pêndulo 1")
xlabel("t [s]")
ylabel("theta [rad]")
plot(T,Y1,'b')
show()

title("Espaço de fase - Pêndulo 1")
ylabel("w [rad/s]")
xlabel("theta [rad]")
plot(Y1,X1,'b')
show()
```

Neste caso, logo após o pêndulo ser solto do repouso em $\theta(0) = 0.2$ ele começa a sofrer a ação da força de atrito, o que faz o sistema começar a perder energia, fazendo o parâmetro $\theta$ tender a zero em uma primira parte do movimento. Entretando, por causa da força externa que aumenta a energia do sistema, a energia total do sistema se balanceia, fazendo o pêndulo entrar em uma órbita estável após o transiente inicial. Nesta situação, a orbita estavel ocorrer para tempos, aproximadamente, maiores que t = 10, ela também pode ser observada claramente no diagrama de fase. Posteriormente vai ser mostrado que este comportamento estável só é possível por causa do valor de $F_D$, para outros valores o comportamento pode ser bastante diferente. 

**Regime não-caótico:**

O caso apresentado anteriormente é um exemplo de um pêndulo em um regime não-caótico, este tipo de regime se caracteriza pela propriedade que independente de suas condições iniciais ($\theta$ e $w$) ele vai tender a uma mesma órbita estável. Este tipo de comportamento é chamado de não-caótico justamente porque o comportamento futuro do sistema (depois de varias iterações) é completamente previsível. Para visualizar claramente esta propriedade vamos fazer a simulação de dois pêndulos físicos, eles vão ser idênticos (mesmas constantes $q = 1/2$, $l = g = 9.8$, $F_D = 0.5$ e $D = 2/3$), a velocidade angular inicial dos dois também será a mesma $w(0) = 0$, entretando a posição angular inicial de cada um será bem diferente. O pêndulo 2 vai começar em $\theta(0) = 0.2$ e o pêndulo 3 vai começar em $\theta(0) = -\pi/2$. Alem de plotar o $\theta$ em função do tempo ($\theta-t$), e o espaço de fase ($w-\theta$) para cada pêndulo, também vamos plotar a diferença entre os angulos dos dois pêndulos em função do tempo ($\Delta\theta-t$).

```python
l = 9.8
D = 2/3
q = 0.5
Fd = 0.5
a = 0.0
b = 50.0
N = 10**4
h = (b-a)/N

w2,the2,w3,the3 = 0,0.2,0,-pi/2
r2,r3 = array([w2,the2]), array([w3,the3])
X2 = []
Y2 = []
X3 = []
Y3 = []
T = []

for t in linspace(a,b,N):
    if r2[1]>pi:
        r2[1]-=2*pi
    if r2[1]<=-pi:
        r2[1]+=2*pi
    if r3[1]>pi:
        r3[1]-=2*pi
    if r3[1]<=-pi:
        r3[1]+=2*pi
    X2.append(r2[0])
    Y2.append(r2[1])
    X3.append(r3[0])
    Y3.append(r3[1])
    r2,r3 = RK4(r2,t), RK4(r3,t)
    T.append(t)
    
title("Pêndulo 2 e 3")
xlabel("t [s]")
ylabel("theta [rad]")
plot(T,Y2,'r')
plot(T,Y3,'g')
show()

title("Espaço de fase - Pêndulo 2 e 3")
ylabel("w [rad/s]")
xlabel("theta [rad]")
plot(Y2,X2,'r')
plot(Y3,X3,'g')
show()

title("Diferença entre os thetas - Pêndulo 2 e 3")
xlabel("t [s]")
ylabel("dtheta [rad]")
Y2 = array(Y2)     #Tranformar em array
Y3 = array(Y3)
Z = abs(Y3-Y2)
plot(T,Z,'k')
show()
```

Veja que este caso, independente das condições iniciais ambos os pêndulos 2 e 3 tendem a mesma órbita, esta é definida exclusivamente pelas constantes escolhidas inicialmente para o sistema. Perceba também que a diferença entre os thetas $\Delta\theta$ tende a zero com o tempo, isso só reforça que ambos os pêndulos, para tempos grandes, viram um mesmo oscilador harmônico.

**Regime caótico:**

Neste caso, assim como o anterior vamos considerar dois pêndulos identicos (com as mesmas constantes do sistema), entretanto desta vezes vamos aumentar o valor da intensidade da fora externa sob os sistemas para $F_{D} = 1.2$. As outras constantes não vão ter seus valores alterados. Sobre as condições iniciais do sistema, vamos considerar novamente $w(0) = 0$, mas desta vez vamos dar posições angulares iniciais bem próximas para os dois pêndulos, o pêndulo 4 vai começar seu movimento em $\theta(0) = 0.2$ e o pêndulo 5 vai começar com uma diferença de $d = 0.001$ com relação ao pêndulo 4. Desta forma, ambos os sistemas inicialmente são praticamente os mesmos, apenas diferindo infimamente suas posições angulares iniciais. 

```python
Fd = 1.2
d = 0.001
a = 0.0
b = 80.0
N = 5*10**4
h = (b-a)/N

w4,the4,w5,the5 = 0,0.2,0,0.2-d
r4,r5 = array([w4,the4]), array([w5,the5])
X4 = []
Y4 = []
X5 = []
Y5 = []
T = []

for t in linspace(a,b,N):
    if r4[1]>pi:
        r4[1]-=2*pi
    if r4[1]<=-pi:
        r4[1]+=2*pi
    if r5[1]>pi:
        r5[1]-=2*pi
    if r5[1]<=-pi:
        r5[1]+=2*pi
    X4.append(r4[0])
    Y4.append(r4[1])
    X5.append(r5[0])
    Y5.append(r5[1])
    r4,r5 = RK4(r4,t), RK4(r5,t)
    T.append(t)
    
title("Pêndulo 4 e 5")
xlabel("t [s]")
ylabel("theta [rad]")
plot(T,Y4,'c,')
plot(T,Y5,'m,')
show()

title("Espaço de fase - Pêndulo 4 e 5")
ylabel("w [rad/s]")
xlabel("theta [rad]")
plot(Y4,X4,'c,')
plot(Y5,X5,'m,')
show()

title("Diferença entre os thetas - Pêndulo 4 e 5")
xlabel("t [s]")
ylabel("dtheta [rad]")
Y4 = array(Y4)     #Tranformar em array
Y5 = array(Y5)
Z = abs(Y5-Y4)
plot(T,Z,'k,')
show()
```

Perceba que neste caso, por mais que os pêndulos tenham sido definidos inicialmente pelas mesmas contantes, e mesmo que eles tenham começado seu movimento bem próximos um do outro, suas tragetórias finais são bem diferentes. Isto significa que para este modelo com $F_{D} = 1.2$, as condições iniciais de um pêndulo são extremamente importantes para definir o movimento dele a longo prazo, diferente do regime não-caótico. É possivel perceber olhando para o plot ($\Delta\theta-t$), que com o passar do tempo a diferença entre os thetas nunca vai tender a zero, indicando que realmente os pêndulos 4 e 5 não são atraidos para um mesmo movimento. Por mais que tenhamos simulado apenas o tempo entre 0.0 e 80.0, o movimento dos pêndulos para tempos maiores também apresentam as mesmas características, ambos nunca mais terão o mesmo movimento novamente. Este fenômeno ocorre em sistemas caóticos e é chamado de "sensibilidade as condições inicias". 

**Seção de Poincaré e o atrator estranho:**

Observando os resultados encontrados anteriormente para os espaço de fase dos pêndulos 4 e 5, podemos perceber que por mais que os sistemas sejam caóticos, as estruturas formadas ainda apresentam certa regularidade. Existe, por exemplo, a formação de órbitas que são bem próximas umas das outras, e que persistem por dois ou três ciclos antes de se desfazerem. Uma das estruturas interessantes que podem aparecer neste tipo de espaço de fase, ocorre quando a força externa aplicada ao pêndulo entra em fase, ou seja quando $Dt = 2\pi$n, sendo $n$ um inteiro. Para analisarmos como esta estrutura se comporta em um espaço de fase ($w-\theta$), vamos usar a secção de Poincaré, um método usado para discretizar um sistema contínuo, particularmente vamos discretizar um pêndulo físico caótico em intervalos onde a força externa esteja aproximadamente em fase.

Para fazer isso, vamos considerar intervalos de $h$ centrados em cada $t = \dfrac{2\pi n}{D}$, já que $h$ é o passo de tempo do nosso programa. Quando o tempo $t$ entrar em um destes intervalos ele será considerado como um dos tempos em que a força externa está em fase. Para fins computacionais, vamos considerar intervalos da forma $\dfrac{2\pi n}{D}-\dfrac{h}{2} < t < \dfrac{2\pi n}{D}+\dfrac{h}{2}$. Vamos usar os mesmos valores de constantes e condições iniciais do pêndulo 5 para fazer o plot do espaço de fase. 

```python
a = 0.0
b = 2500.0
N = 10**5
h = (b-a)/N

r5 = array([w5,the5])
X5 = []
Y5 = []
T = []
n = 0
for t in linspace(a,b,N):
    if r5[1]>pi:
        r5[1]-=2*pi
    if r5[1]<=-pi:
        r5[1]+=2*pi
        
    if t > 2*(n*pi)/D +h/2:
        n+=1
    if (t < 2*(n*pi)/D +h/2) and (t > 2*(n*pi)/D -h/2):
        X5.append(r5[0])
        Y5.append(r5[1])
    T.append(t)
    r5 = RK4(r5,t)
    
title("Atrator estranho - Fd = 1.2")
ylabel("w")
xlabel("theta")
plot(Y5,X5,'.')
show()
```

Esta estrutura criada com $F_{D} = 1.2$, é conhecida como um atrator estranho. Mesmo este sendo um sistema caótico, independentemente das condições iniciais do sistema, todos os pêndulos terão suas orbitas no espaço de fase passando por este atrator quando a força externa estiver em fase. Isto significa que se tivessemos usado o pêndulo 4 no lugar do 5, a figura do atrator seria a mesma. Na verdade este atrator existe tanto para sistemas caóticos quanto para sistemas não-caóticos, sua forma sendo diferente dependendo do valor de $F_{D}$.

**Rotas para o caos:**

Pudemos observar que o valor de $F_D$ determina se o sistema é caótico ou não, mas como se dá a passagem de um regime para outro? Para começarmos a entender isso vamos analisar três pendulos físicos, todos com as mesmas constantes do caso anterior, mas cada um respectivamente com $F_D = 1.35$, $F_D = 1.44$ e $F_D = 1.465$. Suas condições iniciais são $w(0) = 0$ e $\theta(0) = 0.2$.

```python
a = 0.0
b = 100.0
N = 10**4
h = (b-a)/N

w6,the6,w7,the7,w8,the8 = 0.0,0.2,0.0,0.2,0.0,0.2

r6,r7,r8 = array([w6,the6]), array([w7,the7]), array([w8,the8])
Y6 = []
Y7 = []
Y8 = []
T = []

for t in linspace(a,b,N):
    if r6[1]>pi:
        r6[1]-=2*pi
    if r6[1]<=-pi:
        r6[1]+=2*pi
    if r7[1]>pi:
        r7[1]-=2*pi
    if r7[1]<=-pi:
        r7[1]+=2*pi
    if r8[1]>pi:
        r8[1]-=2*pi
    if r8[1]<=-pi:
        r8[1]+=2*pi
    Fd = 1.35
    r6 = RK4(r6,t)
    Y6.append(r6[1])
    
    Fd = 1.44
    r7 = RK4(r7,t)
    Y7.append(r7[1])
    
    Fd = 1.465
    r8 = RK4(r8,t)
    Y8.append(r8[1])
    T.append(t)
    
title("Pêndulo 6 - Fd = 1.35")
xlabel("t [s]")
ylabel("theta [rad]")
plot(T,Y6)
grid()
show()

title("Pêndulo 7 - Fd = 1.44")
xlabel("t [s]")
ylabel("theta [rad]")
plot(T,Y7)
grid()
show()

title("Pêndulo 8 - Fd = 1.465")
xlabel("t [s]")
ylabel("theta [rad]")
plot(T,Y8)
grid()
show()
```

Perceba que conforme nós vamos aumentando o valor de $F_D$ o perido dos pêndulos vai duplicando de tamanho. Para o pêndulo 6 o periodo é igual ao da força externa $T = \dfrac{2\pi}{D} = 3\pi$, já no pêndulo 7 a amplitude é modificada, fazendo com que o período seja 2x o periodo da força externa, no pêndulo 8 este periodo vira 4x. Isso significa que cada vez demora mais para o pêndulo repetir o mesmo movimento, quanto maior este periodo mais próximo o pêndulo estará do caos. Para melhor visualizar esta transição vamos construir um diagrama de bifurcação ($\theta-F_D$). Os valores de $F_D$ entarão entre $1.35$ e $1.49$, intervalo este dividido em 200 partes. Para cada valor de $F_D$ será resolvida uma EDO do pêndulo físico, somente serão salvos os valores de theta quando a força externa estiver em fase, e esta coleta ocorrerá entre o passo 300 e 400 do programa, para que descartemos o transiente inicial. As constantes e condições inicias vão ser as mesmas usadas para resolver o caso anterior.

```python
a = 0.0
b = 3*pi*400
N = 10**4
h = (b-a)/N

w,te = 0,0.2
r = array([w,te])
Y = []    #Theta 
FD = []

c = 1.35
d = 1.49
e = 200     #Partes de Fd
for Fd in linspace(c,d,e):
    t = 0
    r = array([w,te])
    n = 0
    while t<3*pi*400:
        if r[1]>pi:
            r[1]-=2*pi
        if r[1]<=-pi:
            r[1]+=2*pi
        if t > 2*(n*pi)/D +h/2:
            n+=1
        if t>3*pi*300: 
            if (t < 2*(n*pi)/D +h/2) and (t > 2*(n*pi)/D -h/2):
                Y.append(r[1])
                FD.append(Fd)
        r = RK4(r,t) 
        t+=h
title("Diagrama de bifurcação")
xlim(1.35,1.5)
ylim(1,3)
ylabel("theta [rad]")
xlabel("Fd")
plot(FD,Y,',')
show()
```

Neste diagrama de bifurcação podemos identificar claramente o que aconteceu na transição entre os pêndulos 6, 7 e 8. Para pêndulos com $F_D < 1.424$ (como o pêndulo 6) seus periodos são $1xT$, e eles apresentam apenas um valor de $\theta$ para cada $F_D$. No intervalo $1.424 < F_D < 1.459$ os periodos dos pêndulos são $2xT$, e para um mesmo $F_D$ os valores de $\theta$ oscilam entre dois pontos, e assim por diante. Logo, é possivel perceber que quando o diagrama se dividir em n partes o periodo do pêndulo será $nxT$. Quanto mais o valor de $F_D$ cresce mais o diagrama se bifurca e maior é o período do pêndulo, até que o seu movimento não se repita mais e ele entre em um regime caótico.
